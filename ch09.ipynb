混合ガウス分布のEMアルゴリズム
==

Old Faithful 間欠泉データ
--

```python
f = open('old_faithful_data.txt')
data = np.array(list(map(float, f.read().split())))
eruptions = np.array(data[1::3])
waiting = np.array(data[2::3])

eruptions -= eruptions.mean()
eruptions /= eruptions.std()
waiting -= waiting.mean()
waiting /= waiting.std()

plt.figure(figsize=(4, 4))
plt.plot(eruptions, waiting, '.g')
plt.xlim(-2.5, 2.5)
plt.ylim(-2.5, 2.5)
plt.title('Old Faithful 間欠泉データ')
```

K-meansクラスタリング
--

```python
X = np.hstack((eruptions.reshape((-1, 1)), waiting.reshape((-1, 1))))

def k_means(X, n_cluster=2, n_iter=100):
    """ K-meansクラスタリングする """
    N = X.shape[0]
    D = X.shape[-1]
    
    myu = np.zeros((n_cluster, D))
    for k in range(n_cluster):
        myu[k] = X[np.random.randint(N)]
        
    d = np.zeros((len(X), n_cluster))
    for _ in range(n_iter):
        myu_old = myu.copy()
    
        # E-step
        for k in range(n_cluster):
            d[:, k] = np.sum((X - myu[k]) ** 2, axis=1)
        r = np.argmin(d, axis=1)
    
        # M-step
        for k in range(n_cluster):
            myu[k] = X[r == k].mean(axis=0)
        
        opt = np.sum((myu - myu_old) ** 2)
        print(opt)
        if opt < 1e-32:
            break
    
    return myu, r
 
myu_kmeans, r_kmeans = k_means(X)
    
plt.figure(figsize=(4, 4))
plt.plot(X[r_kmeans == 0, 0], X[r_kmeans == 0, 1], '.b')
plt.plot(X[r_kmeans == 1, 0], X[r_kmeans == 1, 1], '.r')
plt.plot(myu_kmeans[0, 0], myu_kmeans[0, 1], '*b', ms=20, markeredgecolor='w', markeredgewidth=2)
plt.plot(myu_kmeans[1, 0], myu_kmeans[1, 1], '*r', ms=20, markeredgecolor='w', markeredgewidth=2)
plt.xlim(-2.5, 2.5)
plt.ylim(-2.5, 2.5)
plt.title('K-means')
```

混合ガウス分布のためのEMアルゴリズム
--

```python
from scipy.stats import multivariate_normal
from matplotlib.patches import Ellipse

def log_likelihood(X, myu, cov, pi):
    """ 対数尤度を得る """
    N = len(X)
    n_cluster = len(myu)
    l = np.zeros((N, n_cluster))
    for k in range(n_cluster):
        var = multivariate_normal(mean=myu[k], cov=cov[k])
        l[:, k] = pi[k] * var.pdf(X)
    return np.sum(np.log(np.sum(l, axis=1)))

def responsibility(X, myu, cov, pi):
    """ 負担率を得る """
    N = len(X)
    n_cluster = len(myu)
    gamm = np.zeros((N, n_cluster))
    for k in range(n_cluster):
        var = multivariate_normal(mean=myu[k], cov=cov[k])
        gamm[:, k] = pi[k] * var.pdf(X)
    for n in range(N):
        gamm[n] /= gamm[n].sum()
    return gamm

def em_for_gaussain_mixtures(X, n_cluster=2, n_iter=100):
    """ 混合ガウス分布のためEMアルゴリズム """
    N = X.shape[0]
    D = X.shape[-1]

    # k-meansで初期値を得る
    myu, r = k_means(X, n_cluster=n_cluster)
    
    pi = np.zeros(n_cluster)
    for k in range(n_cluster):
        pi[k] = float(np.sum(r == k)) / N
    
    cov = np.zeros((n_cluster, D, D))
    for k in range(n_cluster):
        d = X[r == k] - myu[k]
        cov[k] = np.dot(d.T, d) / np.sum(r == k)
     
    opt = log_likelihood(X, myu, cov, pi)
    print(opt)

    d = np.zeros_like(X)
    for _ in range(n_iter):
        opt_old = opt.copy()
        
        # E-step
        gamm = responsibility(X, myu, cov, pi)
        
        # M-step
        for k in range(n_cluster):
            N_k = gamm[:, k].sum()
            myu[k] = np.dot(gamm[:, k], X) / N_k
            
            for n in range(N):
                d[n] = (X[n] - myu[k]) * np.sqrt(gamm[n, k])
            cov[k] = np.dot(d.T, d) / N_k

            pi[k] = N_k / N

        opt = log_likelihood(X, myu, cov, pi)
        print(opt)
        if opt - opt_old < 1e-32:
            break

    return myu, cov, pi

def get_ellipse_param(cov):
    """ 楕円のパラメータを得る """
    x, y = cov[0, 0], cov[1, 1]
    xy = cov[1, 0]
    u = x + y + np.sqrt((x - y) ** 2 + 4 * (xy ** 2)) / 2
    v = x + y - np.sqrt((x - y) ** 2 + 4 * (xy ** 2)) / 2
    ang = np.arctan(u - x / xy) / np.pi * 180
    return u, v, ang

myu_em, cov_em, pi_em = em_for_gaussain_mixtures(X)
gamm = responsibility(X, myu_em, cov_em, pi_em)
r_em = np.argmax(gamm, axis=1)

fig, ax = plt.subplots(figsize=(4, 4))
ax.plot(X[r_em == 0, 0], X[r_em == 0, 1], '.b')
ax.plot(X[r_em == 1, 0], X[r_em == 1, 1], '.r')

u, v, ang = get_ellipse_param(cov_em[0])
ell = Ellipse(xy=myu_em[0], width=u * 2.5, height=v * 2.5, angle=180 - ang, fc='none', ec='b', lw=3)
ax.add_patch(ell)

u, v, ang = get_ellipse_param(cov_em[1])
ell = Ellipse(xy=myu_em[1], width=u * 2.5, height=v * 2.5, angle=180 - ang, fc='none', ec='r', lw=3)
ax.add_patch(ell)

ax.set_xlim(-2.5, 2.5)
ax.set_ylim(-2.5, 2.5)
ax.set_title('混合ガウス分布のためのEMアルゴリズム')
```
