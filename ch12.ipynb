MNIST

```python
im = np.fromfile('mnist/train-images-idx3-ubyte', dtype=np.uint8)[16:].reshape((-1, 28, 28))
label = np.fromfile('mnist/train-labels-idx1-ubyte', dtype=np.uint8)[8:]
im_3 = im[label == 3]
```

主成分分析
--

```python
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(im_3.reshape((-1, 784)))
```

```python
fig, ax = plt.subplots(1, 5, figsize=(15, 4))
ax[0].imshow(pca.mean_.reshape((28, 28)), cmap='gray_r', interpolation='Nearest')
ax[0].axis('off')
ax[0].set_title('平均')
for i in range(4):
    ax[i + 1].imshow(pca.components_[i].reshape((28, 28)), cmap='gray_r', interpolation='Nearest')
    ax[i + 1].axis('off')
    ax[i + 1].set_title(r'$\lambda_{}={:.1e}$'.format(i + 1, pca.explained_variance_[i]))
plt.suptitle('平均と主成分分析の固有ベクトルの画像\n'+'$\lambda$'+'は固有値')
```

次元削減
--

```python
def dimensionality_reduction(pca, im, m):
    """ 次元削減する """
    c = pca.transform(im.reshape((-1, im.shape[0] * im.shape[1])))
    c[:, m:] = 0
    recon = pca.inverse_transform(c).reshape(im.shape)
    return recon
```

```python
fig, ax = plt.subplots(1, 5, figsize=(15, 4))
ax[0].imshow(im_3[0], cmap='gray_r', interpolation='Nearest', vmin=0, vmax=255)
ax[0].axis('off')
ax[0].set_title('真')
for i, m in enumerate([1, 10, 50, 250]):
    recon = dimensionality_reduction(pca, im_3[0], m)
    ax[i + 1].imshow(recon, cmap='gray_r', interpolation='Nearest', vmin=0, vmax=255)
    ax[i + 1].axis('off')
    ax[i + 1].set_title('M={}'.format(m))
plt.suptitle('M個の主成分から再構成した画像')
```

Old Faithful 間欠泉データを線形に前処理
--

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

f = open('old_faithful_data.txt')
data = np.array(list(map(float, f.read().split())))
data = np.c_[np.array(data[1::3]), np.array(data[2::3])]

ss = StandardScaler()
data_ss = ss.fit_transform(data)

pca = PCA()
data_pca = pca.fit_transform(data)
data_w = np.dot(data_pca, np.linalg.pinv(np.sqrt(np.diag(pca.explained_variance_))))

fig, ax = plt.subplots(1, 3, figsize=(12, 4))
ax[0].plot(data[:, 0], data[:, 1], '.')
ax[1].plot(data_ss[:, 0], data_ss[:, 1], '.')
ax[2].plot(data_w[:, 0], data_w[:, 1], '.')
ax[0].set_title('生データ')
ax[1].set_title('標準化')
ax[2].set_title('白色化')
```
